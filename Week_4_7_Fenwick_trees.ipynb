```python
def update(tree, n, index, value):
    while index <= n:
        tree[index] += value
        index += index & -index
def query(tree, index):
    total = 0
    while index > 0:
        total += tree[index]
        index -= index & -index
    return total
n = 5
attendance = [1, 2, 1, 2, 1]
tree = [0] * (n + 1)
for i in range(n):
    update(tree, n, i + 1, attendance[i])
# Query 1
print("Total attendance till Day 4 =", query(tree, 4))
difference = 2 - attendance[2]
attendance[2] = 2
update(tree, n, 3, difference)
# Query 2
print("After update, total attendance till Day 4 =", query(tree, 4))
```
