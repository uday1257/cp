```python
class SegmentTree:
    def __init__(self, arr):
        self.n = len(arr)
        self.arr = arr
        self.min_tree = [0] * (4 * self.n)
        self.max_tree = [0] * (4 * self.n)
        self.build(0, 0, self.n - 1)

    def build(self, node, start, end):
        if start == end:
            self.min_tree[node] = self.arr[start]
            self.max_tree[node] = self.arr[start]
        else:
            mid = (start + end) // 2
            self.build(2*node+1, start, mid)
            self.build(2*node+2, mid+1, end)

            self.min_tree[node] = min(
                self.min_tree[2*node+1],
                self.min_tree[2*node+2]
            )

            self.max_tree[node] = max(
                self.max_tree[2*node+1],
                self.max_tree[2*node+2]
            )

    def range_min(self, l, r, node=0, start=0, end=None):
        if end is None:
            end = self.n - 1

        if r < start or end < l:
            return float('inf')

        if l <= start and end <= r:
            return self.min_tree[node]

        mid = (start + end) // 2

        return min(
            self.range_min(l, r, 2*node+1, start, mid),
            self.range_min(l, r, 2*node+2, mid+1, end)
        )

    def range_max(self, l, r, node=0, start=0, end=None):
        if end is None:
            end = self.n - 1

        if r < start or end < l:
            return float('-inf')

        if l <= start and end <= r:
            return self.max_tree[node]

        mid = (start + end) // 2

        return max(
            self.range_max(l, r, 2*node+1, start, mid),
            self.range_max(l, r, 2*node+2, mid+1, end)
        )

    def update(self, idx, val, node=0, start=0, end=None):
        if end is None:
            end = self.n - 1

        if start == end:
            self.arr[idx] = val
            self.min_tree[node] = val
            self.max_tree[node] = val
        else:
            mid = (start + end) // 2

            if idx <= mid:
                self.update(idx, val, 2*node+1, start, mid)
            else:
                self.update(idx, val, 2*node+2, mid+1, end)

            self.min_tree[node] = min(
                self.min_tree[2*node+1],
                self.min_tree[2*node+2]
            )

            self.max_tree[node] = max(
                self.max_tree[2*node+1],
                self.max_tree[2*node+2]
            )


n = int(input("Number of sensors: "))
arr = list(map(int, input("Temperature readings: ").split()))

st = SegmentTree(arr)

results = []
print("Queries ")
l, r = map(int, input("RangeMin: ").split())
results.append(st.range_min(l, r))

l, r = map(int, input("RangeMax: ").split())
results.append(st.range_max(l, r))

idx, val = map(int, input("Update: ").split())
st.update(idx, val)

l, r = map(int, input("RangeMin: ").split())
results.append(st.range_min(l, r))


for res in results:
    print(res)
```
