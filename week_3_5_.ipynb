```python
from bisect import bisect_left

def subset_sum_meet_in_middle(arr, S):
    n = len(arr)
    mid = n // 2
    left = arr[:mid]
    right = arr[mid:]
    def generate_sums(nums):
        sums = []
        def dfs(i, current_sum):
            if i == len(nums):
                sums.append(current_sum)
                return
            dfs(i + 1, current_sum)           # exclude element
            dfs(i + 1, current_sum + nums[i]) # include element
        dfs(0, 0)
        return sums

    left_sums = generate_sums(left)
    right_sums = generate_sums(right)
    right_sums.sort()
    for ls in left_sums:
        need = S - ls
        idx = bisect_left(right_sums, need)
        if idx < len(right_sums) and right_sums[idx] == need:
            return True
    return False
T = int(input().strip())
for _ in range(T):
    N, S = map(int, input().split())
    arr = list(map(int, input().split()))
    if subset_sum_meet_in_middle(arr, S):
        print("YES")
    else:
        print("NO")
```
